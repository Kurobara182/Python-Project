# 2.c create a new column named age_category whose values should be based on the values in the age column, divide the values into 3 ordinal numbers; 1 – 15 to 17, 2 – 18 to 20, 3 – 21 and over

```python
def new_age(age):
    if age <= 17:
        return 1
    elif age <= 20:
        return 2
    else:
        return 3
    
```

```python
#create the new column age_category and pass the corresponding value from new_age
data['age_category'] = data['age'].apply(new_age)
data[['age', 'age_category']].head(10)
```

```python
def new_passed(G3):
    if  G3 >= 8:
        return 'yes'
    elif G3 < 8:
        return 'no'
   
```

```python
#create the new table passed and pass the corresponding value from new_passed
data['passed'] = data['G3'].apply(new_passed)
data[['G3', 'passed']].head(10)
```
